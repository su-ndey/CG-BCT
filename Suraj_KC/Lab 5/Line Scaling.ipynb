```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def BLA_line(x1,y1,x2,y2):
    dx = abs(x2-x1)
    dy = abs(y2-y1)
    sx = 1 if x2 > x1 else -1
    sy = 1 if y2 > y1 else -1
    x_coordinates ,y_coordinates =[],[]
    if (dx >dy):
        P = (2*dy-dx)
        for _ in range(dx+1):
            x_coordinates.append(x1)
            y_coordinates.append(y1)
            if (P >=0):
                y1 += sy
                P=P + 2*dy - 2*dx
            else:
                P=P + 2*dy
            x1 += sx
        
    else:
        
        P= (2*dx-dy)
        for _ in range(dy+1):
            x_coordinates.append(x1)
            y_coordinates.append(y1)
            y1 += sy
            if P >=0:
                x1 += sx
                P=P + 2*dx -2*dy
            else:
                P=P + 2*dx
    return x_coordinates,y_coordinates
```

```python
x1,y1 = map(int,input("Enter starting coordinates :").split(','))
x2,y2 = map(int,input("Enter final coordinates :").split(','))
sx,sy = map(int,input("Enter scaling factor for x and y :").split(','))
```

```python
xes,yes = BLA_line(x1,y1,x2,y2)
```

```python
t = np.array([[1,0,-x1],[0,1,-y1],[0,0,1]])
s = np.array([[sx,0,0],[0,sy,0],[0,0,1]])
T = np.array([[1,0,x1],[0,1,y1],[0,0,1]])
CM = T@s@t
points = np.vstack((xes,yes,np.ones_like(xes)))
TP = CM@points
TP
```

```python
plt.figure(figsize=(8,5))
plt.plot(xes,yes,marker="o")
plt.plot(TP[0],TP[1],marker="o")
plt.grid()
plt.savefig('output_for_line_scaling.png')
plt.xlabel("X-coordinate")
plt.ylabel("Y_coordinate")
plt.title("Line Scaling")
```
