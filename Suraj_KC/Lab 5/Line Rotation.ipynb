```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def BLA_line(x1,y1,x2,y2):
    dx = abs(x2-x1)
    dy = abs(y2-y1)
    sx = 1 if x2 > x1 else -1
    sy = 1 if y2 > y1 else -1
    x_coordinates ,y_coordinates =[],[]
    if (dx >dy):
        P = (2*dy-dx)
        for _ in range(dx+1):
            x_coordinates.append(x1)
            y_coordinates.append(y1)
            if (P >=0):
                y1 += sy
                P=P + 2*dy - 2*dx
            else:
                P=P + 2*dy
            x1 += sx
        
    else:
        
        P= (2*dx-dy)
        for _ in range(dy+1):
            x_coordinates.append(x1)
            y_coordinates.append(y1)
            y1 += sy
            if P >=0:
                x1 += sx
                P=P + 2*dx -2*dy
            else:
                P=P + 2*dx
    return x_coordinates,y_coordinates
```

```python
x1,y1 = map(int,input("Enter starting coordinates :").split(','))
x2,y2 = map(int,input("Enter final coordinates :").split(','))
theta = np.radians(int(input("Enter angle :")))
```

```python
xes,yes = BLA_line(x1,y1,x2,y2)
```

```python
t = np.array([[1,0,-x1],[0,1,-y1],[0,0,1]])
R = np.array([[np.cos(theta),-np.sin(theta),0],[np.sin(theta),np.cos(theta),0],[0,0,1]])
T = np.array([[1,0,x1],[0,1,y1],[0,0,1]])
CM = T@R@t
points = np.vstack((xes,yes,np.ones_like(xes)))
TP = CM@points
```

```python
plt.plot(xes,yes,marker='o',label="Original Line")
plt.plot(TP[0],TP[1],marker='o',label="Rotated Line")
plt.title("Line Rotation")
plt.xlabel("X-values")
plt.ylabel("Y-values")
plt.savefig("output_for_line_rotation.png")
plt.legend()
plt.grid()
```
