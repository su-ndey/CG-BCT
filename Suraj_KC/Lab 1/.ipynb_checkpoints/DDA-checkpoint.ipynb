```python
import matplotlib.pyplot as plt
```

```python
x1,y1 = map(int, input("Enter starting coordinates: ").split(','))
x2,y2 = map(int,input("Enter final coordinates: ").split(','))
```

```python
dx = x2-x1
dy = y2-y1
steps = max(abs(dx),abs(dy))
x_increment = dx/steps
y_increment = dy/steps
```

```python
x_coordinates,y_coordinates = [],[]
for i in range(int(steps)):
    x1 += x_increment
    y1 += y_increment
    x_coordinates.append(round(x1))
    y_coordinates.append(round(y1))
    print(f"({round(x1)},{round(y1)})")
plt.plot(x_coordinates,y_coordinates,marker='o',markersize=5,color="red")
plt.title("DDA Algorithm")
plt.xlabel("X-coordinate")
plt.ylabel("Y-coordinate")
plt.grid()
plt.savefig('line-graph.png')
plt.show()
```
