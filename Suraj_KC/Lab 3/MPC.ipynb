```python
import matplotlib.pyplot as plt
```

```python
xc,yc = map(int,input("Enter centre coordinates :").split(','))
r = int(input("Enter radius :"))
```

```python
x=0
y=r
p = 1-r
xes =[]
yes = []
```

```python
def symmetry_plot(xes,yes,x,y,xc,yc):
    xes.extend([x,-x,-x,x,y,-y,y,-y])
    yes.extend([y,y,-y,-y,x,x,-x,-x])
```

```python
symmetry_plot(xes,yes,x,y,xc,yc)
while x<y:
    x+=1
    if p<0:
        p = p+2*x+1
    else:
        y -=1
        p = p+2*(x-y) +1
    symmetry_plot(xes,yes,x,y,xc,yc)
```

```python
plt.scatter(xes,yes,color="red")
plt.gca().set_aspect('equal',adjustable='box')
plt.title("Mid-point Circle Drawing Algorithm")
plt.grid()
plt.xlabel("X values")
plt.ylabel("Y values")
plt.savefig('output.png')
```
