```python
import matplotlib.pyplot as plt
```

```python
def BLA_line(x1,y1,x2,y2):
    dx = abs(x2-x1)
    dy = abs(y2-y1)
    sx = 1 if x2 > x1 else -1
    sy = 1 if y2 > y1 else -1
    x_coordinates ,y_coordinates =[],[]
    if (dx >=dy):
        P = (2*dy-dx)
        for _ in range(dx+1):
            x_coordinates.append(x1)
            y_coordinates.append(y1)
            if (P >=0):
                y1 += sy
                P=P + 2*dy - 2*dx
            else:
                P=P + 2*dy
            x1 += sx
        
    else:
        P= (2*dx-dy)
        for _ in range(dy+1):
            x_coordinates.append(x1)
            y_coordinates.append(y1)
            y1 += sy
            if P >=0:
                P=P + 2*dx -2*dy
            else:
                P=P + 2*dx
        
        
    plt.plot(x_coordinates,y_coordinates,marker='o',color="red")
    plt.grid()
    plt.title("Bresenham's Line Drawing Algorithm")
    plt.xlabel("X values")
    plt.ylabel("Y values")
    for _ in range(len(x_coordinates)):
        plt.text(x_coordinates[_],y_coordinates[_],f"({x_coordinates[_]},{y_coordinates[_]})")
        
```

```python
x1,y1 = map(int,input("Enter starting coordinates :").split(','))
x2,y2 = map(int,input("Enter final coordinates :").split(','))
```

```python
BLA_line(x1,y1,x2,y2)
```
