### LAB-1 DDA

```python
#digital dynamic analyser 
import matplotlib.pyplot as plt


def dda_line(x1, y1, x2, y2):
 
    dx = x2 - x1
    dy = y2 - y1
    
 
    steps = max(abs(dx), abs(dy))
    

    x_increment = dx / steps
    y_increment = dy / steps

    x = x1
    y = y1
    
    points = [(x, y)]
    
  
    for _ in range(int(steps)):
        x += x_increment
        y += y_increment
        points.append((round(x), round(y))) 
    
    return points

x1, y1 = 0, 0  
x2, y2 = 8, 16

points = dda_line(x1, y1, x2, y2)
x_coords, y_coords = zip(*points)


plt.plot(x_coords, y_coords, marker='x', color='blue')
plt.title("Digital Line Drawing Algorithm")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.show()
```

### LAB-3 MID-POINT CIRCLE DRAWING ALGORITHM

```python
import matplotlib.pyplot as plt
r=(int(input('enter r')))
xc=(int(input('enter center xc:')))
yc=(int(input('enter center yc:')))

x=0
y=r

p=1-r

xes=[x+xc]
yes=[y+yc]

xes.append(xc+x)
yes.append(yc+y)

xes.append(xc-x)
yes.append(yc+y)

xes.append(xc-x)
yes.append(yc-y)

xes.append(xc+x)
yes.append(yc-y)

xes.append(xc+y)
yes.append(yc+x)

xes.append(xc-y)
yes.append(yc+x)

xes.append(xc+y)
yes.append(yc-x)

xes.append(xc-y)
yes.append(yc-x)
while x<y:
    x=x+1
    if p<0:
        p=p+2*x+1
    if p>=0:
        y=y-1
        p=p+2*(x-y)+1

    xes.append(xc+x)
    yes.append(yc+y)
    
    xes.append(xc-x)
    yes.append(yc+y)
    
    xes.append(xc-x)
    yes.append(yc-y)
    
    xes.append(xc+x)
    yes.append(yc-y)
    
    xes.append(xc+y)
    yes.append(yc+x)
    
    xes.append(xc-y)
    yes.append(yc+x)
    
    xes.append(xc+y)
    yes.append(yc-x)
    
    xes.append(xc-y)
    yes.append(yc-x)

plt.scatter(xes,yes,marker='x')
plt.title('MID-POINT')
plt.show()
```

### LAB-2 BRESSENHAM'S LINE ALGORITHM

```python
import matplotlib.pyplot as plt

x1 = int(input('enter x1:'))
y1 = int(input('enter y1:'))
x2 = int(input('enter x2:'))
y2 = int(input('enter y2:'))
dx=x2-x1
dy=y2-y1
p = 2*dy-dx 

xplot= [x1]
yplot= [y1]
while x1 <= x2:
  xplot.append(x1)
  yplot.append(y1)
  if p<0:
    x1=x1+1
    p = p+2*dy
  else:
    x1=x1+1
    y1=y1+1
    p = p+2*dy-2*dx
    print(f'{x1},{y1}')

plt.plot(xplot,yplot,marker='*')   

plt.show()
```
