### LAB-4 (MID-POINT ELLIPSE)

```python
import matplotlib.pyplot as plt
def midpoint_ellipse(rx,ry,xc,yc):
    x=0
    y=ry
    
    p1 = (ry*ry) - (rx*rx * ry) + (0.25 * rx*rx)

    xes=[]
    yes=[]
    while (2*ry*ry*x <= 2*rx*rx*y):
        if (p1 < 0):
            x += 1
            p1 += 2*ry*ry*x+ry*ry
        else:
            x += 1
            y -= 1
            p1=p1+2*ry*ry*x-2*rx*rx*y+ry*ry
        xes.append(xc+x)
        yes.append(yc+y)
        xes.append(xc-x)
        yes.append(yc+y)
        xes.append(xc+x)
        yes.append(yc-y) 
        xes.append(xc-x)
        yes.append(yc-y)

    p2 = (ry*ry*(x + 0.5) ** 2)+((rx*rx)*(y - 1)**2) - (rx*rx*ry*ry)

    while (y >=0):
        if p2>0:
            y-=1
            p2-=(2*rx*rx*y+rx*rx)
        else:
            x+=1
            y-=1
            p2 += (2*ry*ry * x - 2*rx*ry * y + rx*rx)
        
        xes.append(x+xc)
        yes.append(y+yc)
        xes.append(-x+xc)
        yes.append(y+yc)
        xes.append(x+xc)
        yes.append(-y+yc)
        xes.append(-x+xc)
        yes.append(-y+yc)
    plt.scatter(xes,yes,marker='*')
    plt.show()
rx=int(input("enter rx:"))
ry=int(input("enter ry:"))
xc=int(input("enter xc:"))
yc=int(input("enter yc:"))
midpoint_ellipse(rx,ry,xc,yc)
```
