# Algorithm for Drawing and Scaling a Circle

## Step 1: Input Circle Parameters
- Accept the x-coordinate (`xc`) and y-coordinate (`yc`) of the circle's center.
- Accept the radius (`r`) of the circle.

## Step 2: Initialize Circle Variables
- Set the initial point `(x, y)` to `(0, r)`.
- Initialize the decision parameter `p` to `1 - r`.

## Step 3: Prepare Data Structures
- Create empty lists `xes` and `yes` to store circle points.

## Step 4: Define Helper Function
- Write a function `add_circle_points(xc, yc, x, y)` to compute and add symmetric points of the circle to `xes` and `yes`.

## Step 5: Add Initial Points
- Use the helper function to add points corresponding to `(0, r)`.

## Step 6: Midpoint Circle Algorithm
- While `x < y`:
  - Increment `x` by 1.
  - If `p < 0`:
    - Update `p` by adding `2 * x + 1`.
  - Otherwise:
    - Decrement `y` by 1.
    - Update `p` by adding `2 * x - 2 * y + 1`.
  - Add new symmetric points to `xes` and `yes`.

## Step 7: Sort and Organize Points
- Sort the points by their coordinates and return `x_list`, `y_list`, `xc`, and `yc`.

## Step 8: Define Transformation Matrices
- **Translation to Origin:**  
  Create a matrix to translate a point to the origin:
  \[
  \text{Translation to Origin} =
  \begin{bmatrix}
  1 & 0 & -tx \\
  0 & 1 & -ty \\
  0 & 0 & 1
  \end{bmatrix}
  \]
- **Scaling Matrix:**  
  Create a matrix to scale the circle by `sx` and `sy`:
  \[
  \text{Scaling Matrix} =
  \begin{bmatrix}
  sx & 0 & 0 \\
  0 & sy & 0 \\
  0 & 0 & 1
  \end{bmatrix}
  \]
- **Translation Back from Origin:**  
  Create a matrix to reverse the translation to the origin:
  \[
  \text{Translation Back from Origin} =
  \begin{bmatrix}
  1 & 0 & tx \\
  0 & 1 & ty \\
  0 & 0 & 1
  \end{bmatrix}
  \]

## Step 9: Get Circle Points
- Call the `circle()` function to generate the original circle points.

## Step 10: Input Transformation Parameters
- Set translation offsets (`tx`, `ty`) to `xc` and `yc`.
- Define the scaling factors (`sx`, `sy`).

## Step 11: Compute Composite Transformation Matrix
- Multiply the translation to origin, scaling, and translation back matrices in order:  
  \[
  \text{Composite Matrix} = \text{Translation Back} \cdot \text{Scaling Matrix} \cdot \text{Translation to Origin}
  \]

## Step 12: Apply Transformation
- Stack `x_list` and `y_list` into a matrix of homogeneous coordinates.
- Apply the composite transformation matrix to the points.

## Step 13: Extract Transformed Points
- Extract the x and y coordinates from the transformed matrix.

## Step 14: Visualize Results
- Plot the original circle points in blue.
- Plot the transformed circle points in red.
- Add labels, grid, and legend to the plot.
