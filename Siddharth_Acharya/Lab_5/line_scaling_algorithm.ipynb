# Bresenham's Line Algorithm with Scaling Transformation

## Step 1: Input Line Parameters
- **User Input:** Accept starting `(x1, y1)` and ending `(x2, y2)` coordinates for the line.

## Step 2: Calculate Differences
- Compute the absolute differences `dx = |x2 - x1|` and `dy = |y2 - y1|`.

## Step 3: Determine Step Directions
- Set `sx` and `sy` based on whether `x2 > x1` and `y2 > y1`, to determine the direction for each axis.

## Step 4: Bresenham's Line Algorithm
- If the line has a shallow slope (|m| â‰¤ 1), iterate along the x-axis and adjust the y-coordinate based on the decision parameter.
- If the line has a steep slope (|m| > 1), iterate along the y-axis and adjust the x-coordinate based on the decision parameter.

## Step 5: Store Points
- Collect the `(x, y)` points during each iteration.

## Step 6: Return Points
- Extract the x and y coordinates for plotting purposes.

## Step 7: Define Transformation Matrices
- **Translation to Origin:**  
  The matrix for translating the line to the origin:
  \[
  \text{Translation to Origin} =
  \begin{bmatrix}
  1 & 0 & -tx \\
  0 & 1 & -ty \\
  0 & 0 & 1
  \end{bmatrix}
  \]
- **Scaling Matrix:**  
  The matrix to scale the line by factors `sx` and `sy`:
  \[
  \text{Scaling Matrix} =
  \begin{bmatrix}
  sx & 0 & 0 \\
  0 & sy & 0 \\
  0 & 0 & 1
  \end{bmatrix}
  \]
- **Translation Back from Origin:**  
  The matrix for translating the line back after transformation:
  \[
  \text{Translation Back from Origin} =
  \begin{bmatrix}
  1 & 0 & tx \\
  0 & 1 & ty \\
  0 & 0 & 1
  \end{bmatrix}
  \]

## Step 8: Apply Transformation
- Compute the composite transformation matrix as:
  \[
  \text{Composite Matrix} = \text{Translation Back} \cdot \text{Scaling Matrix} \cdot \text{Translation to Origin}
  \]
- Apply this transformation matrix to the points generated by Bresenham's line algorithm.

## Step 9: Plot the Results
- Plot the original line in **blue** and the transformed line in **red**.

- Display a plot showing both the original and transformed lines.
