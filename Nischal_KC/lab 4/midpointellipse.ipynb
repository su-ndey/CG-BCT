```python
import matplotlib.pyplot as plt

def plot_ellipse_points(xc, yc, x, y):
    """
    Plots symmetric points for the ellipse in all four quadrants.
    """
    plt.plot(xc + x, yc + y, 'ro')  # Quadrant 1
    plt.plot(xc - x, yc + y, 'ro')  # Quadrant 2
    plt.plot(xc + x, yc - y, 'ro')  # Quadrant 3
    plt.plot(xc - x, yc - y, 'ro')  # Quadrant 4

def bresenham_ellipse(xc, yc, rx, ry):
    """
    Bresenham's Midpoint Ellipse Algorithm to draw an ellipse with center (xc, yc) and radii rx, ry.
    """
    # Region 1 Variables
    x = 0
    y = ry
    rx2 = rx * rx
    ry2 = ry * ry
    two_rx2 = 2 * rx2
    two_ry2 = 2 * ry2

    # Initial decision parameter for Region 1
    p1 = ry2 - (rx2 * ry) + (0.25 * rx2)

    # Plot initial points
    plot_ellipse_points(xc, yc, x, y)

    # Region 1
    while (two_ry2 * x) < (two_rx2 * y):
        x += 1
        if p1 < 0:
            p1 += two_ry2 * x + ry2
        else:
            y -= 1
            p1 += two_ry2 * x - two_rx2 * y + ry2
        plot_ellipse_points(xc, yc, x, y)

    # Region 2 Variables
    p2 = (ry2) * ((x + 0.5) ** 2) + (rx2) * ((y - 1) ** 2) - (rx2 * ry2)

    # Region 2
    while y > 0:
        y -= 1
        if p2 > 0:
            p2 += rx2 - two_rx2 * y
        else:
            x += 1
            p2 += two_ry2 * x - two_rx2 * y + rx2
        plot_ellipse_points(xc, yc, x, y)

# Example Usage
if __name__ == "__main__":
    # Ellipse Parameters
    xc, yc = 0, 0  # Center of the ellipse
    rx, ry = 8, 5  # Radii of the ellipse

    # Draw the ellipse
    plt.figure(figsize=(6, 6))
    plt.title("Bresenham's Midpoint Ellipse Algorithm")
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.grid()
    plt.axhline(0, color='black',linewidth=0.5)
    plt.axvline(0, color='black',linewidth=0.5)

    bresenham_ellipse(xc, yc, rx, ry)
    plt.gca().set_aspect('equal', adjustable='box')
    plt.show()
```
