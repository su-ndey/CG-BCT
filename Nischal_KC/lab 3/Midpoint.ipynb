```python
import matplotlib.pyplot as plt
xc = int(input("Enter the point xc"))
yc = int(input("Enter the point yc"))
r = int(input("Enter the point r"))
p=1-r
xe=[]
ye=[]
x=0
y=r

def symmetry(xe,ye,x,y,xc,yc):
    xe.extend([x,-x,-x,x,y,-y,y,-y])
    ye.extend([y,y,-y,-y,x,x,-x,-x])
     
symmetry(xe,ye,x,y,xc,yc)
while x<y:
    x=x+1
    if p<0:
        p=p+2*x+1
    else:
        y=y-1
        p=p+2*(x-y)+1
    symmetry(xe,ye,x,y,xc,yc)
plt.scatter(xe,ye)     
plt.gca().set_aspect('equal',adjustable='box')
```
