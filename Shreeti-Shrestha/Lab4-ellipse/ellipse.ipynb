```python
import matplotlib.pyplot as plt

rx=int(input("Enter coordinate of x axis:"))
ry=int(input("Enter coordinate of y axis:"))
xc=int(input("Enter x coordinate of center:"))
yc=int(input("Enter y coordinate of center:"))

x=0
y=ry


p1=ry**2-(rx**2*ry) + (0.25*rx**2)

xes = []
yes = []

xes.append(x + xc)
yes.append(y + yc)
xes.append(-x + xc)
yes.append(y + yc)
xes.append(x + xc)
yes.append(-y + yc)
xes.append(-x + xc)
yes.append(-y + yc)

#for region 1
while(2*ry**2*x<=2*rx**2*y):
    if (p1<0):
        x=x+1
        p1=p1+2*ry**2*x+ry**2
    else:
        x=x+1
        y=y-1
        p1=p1+ry**2*x-2*rx**2*y+ry**2       
    xes.append(x + xc)
    yes.append(y + yc)
    xes.append(-x + xc)
    yes.append(y + yc)
    xes.append(x + xc)
    yes.append(-y + yc)
    xes.append(-x + xc)
    yes.append(-y + yc)
#for region 2
p2=(ry**2*(x+0.5)**2)+(rx**2*(y-1)**2)-(rx**2*ry**2)
while(y>=0):
    if (p2>0):
         y=y-1
         p2=p2-2*rx**2*y+rx**2
    else:
          x=x+1
          y=y-1
          p2=p2+2*ry**2*x-2*rx**2*y+rx**2
    xes.append(x + xc)
    yes.append(y + yc)
    xes.append(-x + xc)
    yes.append(y + yc)
    xes.append(x + xc)
    yes.append(-y + yc)
    xes.append(-x + xc)
    yes.append(-y + yc)
plt.figure(figsize=(8, 6)) 
plt.scatter(xes,yes,marker=".", color ="green")
plt.title("Midpoint Ellipse Drawing Algorithm")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()
```
